```python
# =============================================================================
#  TechStore - Sistema de Gestión de Ventas
#  Proyecto Final - Estructuras de Datos
# =============================================================================

import csv
import os
from datetime import datetime

# =============================================================================
# DATOS GLOBALES
# =============================================================================

ventas = []          # Lista principal de registros (arreglo dinámico)
_id_counter = 1      # Contador auto-incremental de IDs


# =============================================================================
# MÓDULO 0 – UTILIDADES INTERNAS
# =============================================================================

def _next_id():
    global _id_counter
    uid = _id_counter
    _id_counter += 1
    return uid


def _separador(char="=", ancho=70):
    print(char * ancho)


def _titulo(texto):
    _separador()
    print(f"  {texto}")
    _separador()


def _pausa():
    input("\n  Presiona Enter para continuar...")


# =============================================================================
# MÓDULO 0B – VALIDACIÓN DE ENTRADAS
# =============================================================================

def validar_fecha(texto):
    """Valida que la fecha tenga formato DD/MM/AAAA. Retorna True/False."""
    try:
        datetime.strptime(texto.strip(), "%d/%m/%Y")
        return True
    except ValueError:
        return False


def pedir_float(mensaje, minimo=0.01):
    """Pide un número flotante mayor que `minimo`. Reintenta hasta lograrlo."""
    while True:
        try:
            valor = float(input(mensaje))
            if valor < minimo:
                print(f"  ✗ El valor debe ser mayor a {minimo}. Intenta de nuevo.")
            else:
                return valor
        except ValueError:
            print("  ✗ Entrada inválida. Escribe un número (ej: 150.50).")


def pedir_int(mensaje, minimo=1):
    """Pide un entero >= `minimo`."""
    while True:
        try:
            valor = int(input(mensaje))
            if valor < minimo:
                print(f"  ✗ El valor debe ser al menos {minimo}.")
            else:
                return valor
        except ValueError:
            print("  ✗ Entrada inválida. Escribe un número entero.")


def pedir_texto(mensaje, max_len=50, puede_vacio=False):
    """Pide un texto no vacío con longitud máxima."""
    while True:
        valor = input(mensaje).strip()
        if not puede_vacio and valor == "":
            print("  ✗ Este campo no puede estar vacío.")
        elif len(valor) > max_len:
            print(f"  ✗ Máximo {max_len} caracteres (ingresaste {len(valor)}).")
        else:
            return valor


def pedir_fecha(mensaje):
    """Pide una fecha con formato DD/MM/AAAA."""
    while True:
        valor = input(mensaje).strip()
        if validar_fecha(valor):
            return valor
        print("  ✗ Formato inválido. Usa DD/MM/AAAA (ej: 15/03/2024).")


def pedir_confirmacion(mensaje):
    """Retorna True si el usuario confirma con 's'."""
    resp = input(f"  {mensaje} (s/n): ").strip().lower()
    return resp == "s"


# =============================================================================
# MÓDULO 1 – GESTIÓN DE REGISTROS
# =============================================================================

def cargar_datos_ejemplo():
    """RF-05: Carga automática de 10+ registros al iniciar."""
    global _id_counter
    datos = [
        ("Laptop HP 15",      3, 2500000, "12/01/2024", "Carlos Ruiz",     "Bogotá"),
        ("Mouse Logitech",    10,   85000, "13/01/2024", "Ana Torres",      "Medellín"),
        ("Teclado Mecánico",  5,  320000, "14/01/2024", "Pedro Gómez",     "Cali"),
        ("Monitor Samsung",   2, 1200000, "15/01/2024", "Carlos Ruiz",     "Bogotá"),
        ("Auriculares Sony",  7,  450000, "16/01/2024", "Ana Torres",      "Barranquilla"),
        ("Tablet Lenovo",     4,  980000, "17/01/2024", "María López",     "Bucaramanga"),
        ("Webcam Logitech",   8,  210000, "18/01/2024", "Pedro Gómez",     "Cali"),
        ("SSD Kingston",     15,  350000, "19/01/2024", "María López",     "Medellín"),
        ("Router TP-Link",    6,  175000, "20/01/2024", "Luis Herrera",    "Bogotá"),
        ("Impresora Epson",   3,  680000, "21/01/2024", "Luis Herrera",    "Barranquilla"),
        ("Laptop Dell XPS",   1, 5800000, "22/01/2024", "Carlos Ruiz",     "Bucaramanga"),
        ("Memoria RAM 16GB",  9,  290000, "23/01/2024", "Ana Torres",      "Cali"),
    ]
    for producto, cantidad, precio, fecha, vendedor, sucursal in datos:
        ventas.append({
            "id":        _next_id(),
            "producto":  producto,
            "cantidad":  cantidad,
            "precio":    precio,
            "total":     cantidad * precio,
            "fecha":     fecha,
            "vendedor":  vendedor,
            "sucursal":  sucursal,
        })
    print(f"  ✔ {len(datos)} registros de ejemplo cargados correctamente.")


def ingresar_venta():
    """RF-01: Ingresa una venta con validación completa."""
    _titulo("INGRESAR NUEVA VENTA")
    producto  = pedir_texto("  Producto  : ")
    cantidad  = pedir_int("  Cantidad  : ")
    precio    = pedir_float("  Precio    : $")
    fecha     = pedir_fecha("  Fecha (DD/MM/AAAA): ")
    vendedor  = pedir_texto("  Vendedor  : ")
    sucursal  = pedir_texto("  Sucursal  : ")

    venta = {
        "id":       _next_id(),
        "producto": producto,
        "cantidad": cantidad,
        "precio":   precio,
        "total":    cantidad * precio,
        "fecha":    fecha,
        "vendedor": vendedor,
        "sucursal": sucursal,
    }
    ventas.append(venta)
    print(f"\n  ✔ Venta registrada con ID #{venta['id']}.")
    _pausa()


def mostrar_ventas(lista=None, titulo="LISTADO DE VENTAS"):
    """RF-02: Tabla formateada de ventas en consola."""
    if lista is None:
        lista = ventas
    _titulo(titulo)
    if not lista:
        print("  (No hay registros para mostrar.)")
        _pausa()
        return
    # Encabezado
    print(f"  {'ID':>4}  {'Producto':<22} {'Cant':>4} {'Precio':>12} {'Total':>14}"
          f"  {'Fecha':<12} {'Vendedor':<18} {'Sucursal':<14}")
    _separador("-")
    for v in lista:
        print(f"  {v['id']:>4}  {v['producto']:<22} {v['cantidad']:>4}"
              f"  ${v['precio']:>11,.0f}  ${v['total']:>13,.0f}"
              f"  {v['fecha']:<12} {v['vendedor']:<18} {v['sucursal']:<14}")
    _separador("-")
    print(f"  Total de registros: {len(lista)}")
    _pausa()


def _buscar_por_id_indice(vid):
    """Retorna el índice del registro con ese ID, o -1 si no existe."""
    for i, v in enumerate(ventas):
        if v["id"] == vid:
            return i
    return -1


def modificar_venta():
    """RF-03: Modifica cualquier campo por ID."""
    _titulo("MODIFICAR VENTA")
    vid = pedir_int("  ID de la venta a modificar: ")
    idx = _buscar_por_id_indice(vid)
    if idx == -1:
        print(f"  ✗ No existe ningún registro con ID #{vid}.")
        _pausa()
        return

    v = ventas[idx]
    print(f"\n  Registro actual:")
    print(f"  Producto: {v['producto']} | Cantidad: {v['cantidad']} | "
          f"Precio: ${v['precio']:,.0f} | Fecha: {v['fecha']}")
    print(f"  Vendedor: {v['vendedor']} | Sucursal: {v['sucursal']}")
    print("\n  Campos: 1-Producto  2-Cantidad  3-Precio  4-Fecha  5-Vendedor  6-Sucursal")
    campo = pedir_int("  Elige campo a modificar (1-6): ", minimo=1)

    if campo == 1:
        v["producto"] = pedir_texto("  Nuevo producto: ")
    elif campo == 2:
        v["cantidad"] = pedir_int("  Nueva cantidad: ")
    elif campo == 3:
        v["precio"] = pedir_float("  Nuevo precio: $")
    elif campo == 4:
        v["fecha"] = pedir_fecha("  Nueva fecha (DD/MM/AAAA): ")
    elif campo == 5:
        v["vendedor"] = pedir_texto("  Nuevo vendedor: ")
    elif campo == 6:
        v["sucursal"] = pedir_texto("  Nueva sucursal: ")
    else:
        print("  ✗ Opción inválida.")
        _pausa()
        return

    v["total"] = v["cantidad"] * v["precio"]
    print("  ✔ Registro actualizado.")
    _pausa()


def eliminar_venta():
    """RF-04: Elimina un registro por ID con confirmación."""
    _titulo("ELIMINAR VENTA")
    vid = pedir_int("  ID de la venta a eliminar: ")
    idx = _buscar_por_id_indice(vid)
    if idx == -1:
        print(f"  ✗ No existe ningún registro con ID #{vid}.")
        _pausa()
        return
    v = ventas[idx]
    print(f"  Registro: {v['producto']} | ${v['total']:,.0f} | {v['fecha']}")
    if pedir_confirmacion("¿Confirmas la eliminación?"):
        ventas.pop(idx)
        print("  ✔ Registro eliminado.")
    else:
        print("  Operación cancelada.")
    _pausa()


# =============================================================================
# MÓDULO 5 – RECURSIVIDAD (se define antes del módulo 2 para reutilizarla)
# =============================================================================

def suma_recursiva(lista, idx=0):
    """RF-21: Suma acumulada de totales de forma recursiva."""
    if idx == len(lista):
        return 0
    return lista[idx]["total"] + suma_recursiva(lista, idx + 1)


def conteo_recursivo(lista, umbral, idx=0):
    """RF-22: Cuenta ventas que superan un umbral (recursivo)."""
    if idx == len(lista):
        return 0
    actual = 1 if lista[idx]["total"] > umbral else 0
    return actual + conteo_recursivo(lista, umbral, idx + 1)


def busqueda_recursiva_id(lista, vid, idx=0):
    """RF-23: Búsqueda recursiva por ID con retroceso (backtracking)."""
    if idx == len(lista):
        return None          # No encontrado
    if lista[idx]["id"] == vid:
        return lista[idx]    # Encontrado → retrocede la pila
    return busqueda_recursiva_id(lista, vid, idx + 1)


def factorial_recursivo(n):
    """RF-24: Cálculo de factorial recursivo (demostración didáctica)."""
    if n < 0:
        raise ValueError("Factorial no definido para negativos.")
    if n == 0:
        return 1
    return n * factorial_recursivo(n - 1)


def promedio_recursivo(lista, idx=0, acum=0):
    """Promedio usando recursividad."""
    if idx == len(lista):
        return acum / len(lista) if lista else 0
    return promedio_recursivo(lista, idx + 1, acum + lista[idx]["total"])


# =============================================================================
# MÓDULO 2 – ANÁLISIS ESTADÍSTICO
# =============================================================================

def _totales(lista=None):
    if lista is None:
        lista = ventas
    return [v["total"] for v in lista]


def calcular_estadisticas():
    """RF-06 a RF-11: Despliega todos los indicadores estadísticos."""
    _titulo("ANÁLISIS ESTADÍSTICO")
    if not ventas:
        print("  (Sin registros.)")
        _pausa()
        return

    tots = _totales()
    n = len(tots)

    # RF-06 Total acumulado (iterativo y recursivo)
    total_iter = sum(tots)
    total_rec  = suma_recursiva(ventas)
    print(f"\n  {'─'*60}")
    print(f"  RF-06 │ Total acumulado (iterativo) : ${total_iter:>15,.0f}")
    print(f"         │ Total acumulado (recursivo) : ${total_rec:>15,.0f}")

    # RF-07 Promedio
    prom_iter = total_iter / n
    prom_rec  = promedio_recursivo(ventas)
    print(f"  RF-07 │ Promedio (iterativo)        : ${prom_iter:>15,.0f}")
    print(f"         │ Promedio (recursivo)        : ${prom_rec:>15,.0f}")

    # RF-08 Máximo y mínimo
    v_max = ventas[0]
    v_min = ventas[0]
    for v in ventas[1:]:
        if v["total"] > v_max["total"]:
            v_max = v
        if v["total"] < v_min["total"]:
            v_min = v
    print(f"  RF-08 │ Venta MÁXIMA: ID#{v_max['id']} {v_max['producto']:<20} ${v_max['total']:>12,.0f}")
    print(f"         │ Venta MÍNIMA: ID#{v_min['id']} {v_min['producto']:<20} ${v_min['total']:>12,.0f}")

    # RF-09 Mediana
    tots_ord = sorted(tots)
    if n % 2 == 0:
        mediana = (tots_ord[n // 2 - 1] + tots_ord[n // 2]) / 2
    else:
        mediana = tots_ord[n // 2]
    print(f"  RF-09 │ Mediana                     : ${mediana:>15,.0f}")

    # RF-10 Top 3 productos por cantidad
    conteo = {}
    for v in ventas:
        conteo[v["producto"]] = conteo.get(v["producto"], 0) + v["cantidad"]
    top3 = sorted(conteo.items(), key=lambda x: x[1], reverse=True)[:3]
    print(f"  RF-10 │ Top 3 productos por cantidad:")
    for i, (prod, qty) in enumerate(top3, 1):
        print(f"         │   {i}. {prod:<25} → {qty} unidades")

    # RF-11 Ventas por vendedor
    vend_sum = {}
    vend_cnt = {}
    for v in ventas:
        key = v["vendedor"]
        vend_sum[key] = vend_sum.get(key, 0) + v["total"]
        vend_cnt[key] = vend_cnt.get(key, 0) + 1
    print(f"  RF-11 │ Ventas por vendedor:")
    for vend in vend_sum:
        prom_v = vend_sum[vend] / vend_cnt[vend]
        print(f"         │   {vend:<18} Suma: ${vend_sum[vend]:>12,.0f}  "
              f"Prom: ${prom_v:>12,.0f}  Tx: {vend_cnt[vend]}")

    print(f"  {'─'*60}")
    _pausa()


def demo_recursividad():
    """RF-21 a RF-24: Menú de demostración de funciones recursivas."""
    _titulo("MÓDULO DE RECURSIVIDAD")
    print(f"  RF-21  Suma recursiva de totales   : ${suma_recursiva(ventas):,.0f}")

    umbral = pedir_float("  RF-22  Ingresa umbral para conteo ($): ", minimo=0)
    cnt = conteo_recursivo(ventas, umbral)
    print(f"         Ventas que superan ${umbral:,.0f}    : {cnt}")

    vid = pedir_int("  RF-23  ID a buscar (backtracking)   : ")
    resultado = busqueda_recursiva_id(ventas, vid)
    if resultado:
        print(f"         Encontrado → {resultado['producto']} | ${resultado['total']:,.0f}")
    else:
        print(f"         ID #{vid} no encontrado.")

    n_fact = pedir_int("  RF-24  Número para factorial         : ", minimo=0)
    print(f"         {n_fact}! = {factorial_recursivo(n_fact):,}")
    _pausa()


# =============================================================================
# MÓDULO 3 – ORDENAMIENTO (implementados desde cero)
# =============================================================================

def _copiar_lista():
    return [dict(v) for v in ventas]


# ---------- Burbuja ----------
def burbuja(lista, clave, ascendente=True):
    """RF-12: Burbuja con conteo de intercambios."""
    arr = lista[:]
    n   = len(arr)
    pasos = intercambios = 0
    for i in range(n - 1):
        for j in range(n - 1 - i):
            pasos += 1
            a, b = arr[j][clave], arr[j + 1][clave]
            if ascendente:
                cond = a > b
            else:
                cond = a < b
            if cond:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                intercambios += 1
    return arr, pasos, intercambios


# ---------- Inserción ----------
def insercion(lista, clave, ascendente=True):
    """RF-13: Ordenamiento por inserción."""
    arr = lista[:]
    pasos = intercambios = 0
    for i in range(1, len(arr)):
        actual = arr[i]
        j = i - 1
        while j >= 0:
            pasos += 1
            a, b = arr[j][clave], actual[clave]
            if ascendente:
                cond = a > b
            else:
                cond = a < b
            if cond:
                arr[j + 1] = arr[j]
                intercambios += 1
                j -= 1
            else:
                break
        arr[j + 1] = actual
    return arr, pasos, intercambios


# ---------- Selección ----------
def seleccion(lista, clave, ascendente=True):
    """RF-14: Ordenamiento por selección."""
    arr = lista[:]
    n = len(arr)
    pasos = intercambios = 0
    for i in range(n):
        idx_ext = i
        for j in range(i + 1, n):
            pasos += 1
            if ascendente:
                cond = arr[j][clave] < arr[idx_ext][clave]
            else:
                cond = arr[j][clave] > arr[idx_ext][clave]
            if cond:
                idx_ext = j
        if idx_ext != i:
            arr[i], arr[idx_ext] = arr[idx_ext], arr[i]
            intercambios += 1
    return arr, pasos, intercambios


# ---------- Quicksort recursivo ----------
_qs_pasos = 0
_qs_intercambios = 0

def _quicksort_rec(arr, bajo, alto, clave, ascendente):
    global _qs_pasos, _qs_intercambios
    if bajo < alto:
        pivote_idx = _particionar(arr, bajo, alto, clave, ascendente)
        _quicksort_rec(arr, bajo, pivote_idx - 1, clave, ascendente)
        _quicksort_rec(arr, pivote_idx + 1, alto, clave, ascendente)


def _particionar(arr, bajo, alto, clave, ascendente):
    global _qs_pasos, _qs_intercambios
    pivote = arr[alto][clave]
    i = bajo - 1
    for j in range(bajo, alto):
        _qs_pasos += 1
        if ascendente:
            cond = arr[j][clave] <= pivote
        else:
            cond = arr[j][clave] >= pivote
        if cond:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
            _qs_intercambios += 1
    arr[i + 1], arr[alto] = arr[alto], arr[i + 1]
    return i + 1


def quicksort(lista, clave, ascendente=True):
    """RF-15: Quicksort recursivo."""
    global _qs_pasos, _qs_intercambios
    arr = lista[:]
    _qs_pasos = 0
    _qs_intercambios = 0
    _quicksort_rec(arr, 0, len(arr) - 1, clave, ascendente)
    return arr, _qs_pasos, _qs_intercambios


def menu_ordenamiento():
    """RF-12 a RF-16: Menú de ordenamiento."""
    _titulo("MÓDULO DE ORDENAMIENTO")
    if not ventas:
        print("  (Sin registros.)")
        _pausa()
        return

    print("  1. Burbuja   → por monto total")
    print("  2. Inserción → por fecha")
    print("  3. Selección → por nombre de producto")
    print("  4. Quicksort → por monto total")
    print("  5. Comparativa de todos los algoritmos")
    opc = pedir_int("  Elige (1-5): ", minimo=1)

    orden = pedir_confirmacion("¿Ascendente? (s=ascendente, n=descendente)")

    copia = _copiar_lista()

    if opc == 1:
        res, p, ic = burbuja(copia, "total", orden)
        mostrar_ventas(res, "RESULTADO BURBUJA")
        print(f"  Pasos: {p} | Intercambios: {ic}")
    elif opc == 2:
        res, p, ic = insercion(copia, "fecha", orden)
        mostrar_ventas(res, "RESULTADO INSERCIÓN (por fecha)")
        print(f"  Pasos: {p} | Intercambios: {ic}")
    elif opc == 3:
        res, p, ic = seleccion(copia, "producto", orden)
        mostrar_ventas(res, "RESULTADO SELECCIÓN (A-Z/Z-A)")
        print(f"  Pasos: {p} | Intercambios: {ic}")
    elif opc == 4:
        res, p, ic = quicksort(copia, "total", orden)
        mostrar_ventas(res, "RESULTADO QUICKSORT")
        print(f"  Pasos: {p} | Intercambios: {ic}")
    elif opc == 5:
        # RF-16 Comparativa
        _separador()
        print(f"  {'Algoritmo':<14} {'Clave':<10} {'Pasos':>8} {'Intercambios':>14}")
        _separador("-")
        for nombre, func, clave in [
            ("Burbuja",   burbuja,   "total"),
            ("Inserción", insercion, "fecha"),
            ("Selección", seleccion, "producto"),
            ("Quicksort", quicksort, "total"),
        ]:
            _, p, ic = func(_copiar_lista(), clave, orden)
            print(f"  {nombre:<14} {clave:<10} {p:>8} {ic:>14}")
        _separador()

    _pausa()


# =============================================================================
# MÓDULO 4 – BÚSQUEDA
# =============================================================================

def busqueda_lineal_iter(campo, valor):
    """RF-17: Búsqueda lineal iterativa. Retorna (coincidentes, comparaciones)."""
    resultado = []
    comparaciones = 0
    valor_lower = str(valor).lower()
    for v in ventas:
        comparaciones += 1
        if str(v.get(campo, "")).lower() == valor_lower:
            resultado.append(v)
    return resultado, comparaciones


def busqueda_lineal_rec(campo, valor, idx=0, acum=None, comp=None):
    """RF-19: Búsqueda lineal recursiva."""
    if acum is None:
        acum = []
    if comp is None:
        comp = [0]
    if idx == len(ventas):
        return acum, comp[0]
    comp[0] += 1
    if str(ventas[idx].get(campo, "")).lower() == str(valor).lower():
        acum.append(ventas[idx])
    return busqueda_lineal_rec(campo, valor, idx + 1, acum, comp)


def busqueda_binaria_iter(lista_ord, monto):
    """RF-18: Búsqueda binaria iterativa sobre lista ordenada por total."""
    izq, der = 0, len(lista_ord) - 1
    comparaciones = 0
    resultados = []
    while izq <= der:
        comparaciones += 1
        mid = (izq + der) // 2
        if lista_ord[mid]["total"] == monto:
            # Expandir para capturar duplicados
            resultados.append(lista_ord[mid])
            k = mid - 1
            while k >= 0 and lista_ord[k]["total"] == monto:
                resultados.append(lista_ord[k])
                k -= 1
            k = mid + 1
            while k < len(lista_ord) and lista_ord[k]["total"] == monto:
                resultados.append(lista_ord[k])
                k += 1
            break
        elif lista_ord[mid]["total"] < monto:
            izq = mid + 1
        else:
            der = mid - 1
    return resultados, comparaciones


def busqueda_binaria_rec(lista_ord, monto, izq=None, der=None, comp=None):
    """RF-19: Búsqueda binaria recursiva."""
    if izq is None:
        izq = 0
    if der is None:
        der = len(lista_ord) - 1
    if comp is None:
        comp = [0]
    if izq > der:
        return [], comp[0]
    comp[0] += 1
    mid = (izq + der) // 2
    if lista_ord[mid]["total"] == monto:
        return [lista_ord[mid]], comp[0]
    elif lista_ord[mid]["total"] < monto:
        return busqueda_binaria_rec(lista_ord, monto, mid + 1, der, comp)
    else:
        return busqueda_binaria_rec(lista_ord, monto, izq, mid - 1, comp)


def menu_busqueda():
    """RF-17 a RF-20: Menú de búsqueda."""
    _titulo("MÓDULO DE BÚSQUEDA")
    print("  1. Lineal iterativa  (por ID / producto / vendedor)")
    print("  2. Lineal recursiva  (por ID / producto / vendedor)")
    print("  3. Binaria iterativa (por monto total)")
    print("  4. Binaria recursiva (por monto total)")
    print("  5. Comparativa lineal vs binaria")
    opc = pedir_int("  Elige (1-5): ", minimo=1)

    if opc in (1, 2):
        print("  Campos: id | producto | vendedor")
        campo = pedir_texto("  Campo a buscar: ", max_len=20)
        valor = pedir_texto("  Valor a buscar: ", max_len=50)
        if opc == 1:
            res, comp = busqueda_lineal_iter(campo, valor)
            tipo = "Lineal Iterativa"
        else:
            res, comp = busqueda_lineal_rec(campo, valor)
            tipo = "Lineal Recursiva"
        print(f"\n  [{tipo}] Comparaciones: {comp} | Encontrados: {len(res)}")
        if res:
            mostrar_ventas(res, f"RESULTADOS – {tipo}")
        else:
            print("  No se encontraron coincidencias.")
            _pausa()

    elif opc in (3, 4):
        lista_ord, _, _ = quicksort(_copiar_lista(), "total")
        monto = pedir_float("  Monto exacto a buscar: $", minimo=0)
        if opc == 3:
            res, comp = busqueda_binaria_iter(lista_ord, monto)
            tipo = "Binaria Iterativa"
        else:
            res, comp = busqueda_binaria_rec(lista_ord, monto)
            tipo = "Binaria Recursiva"
        print(f"\n  [{tipo}] Comparaciones: {comp} | Encontrados: {len(res)}")
        if res:
            mostrar_ventas(res, f"RESULTADOS – {tipo}")
        else:
            print("  No se encontraron ventas con ese monto.")
            _pausa()

    elif opc == 5:
        # RF-20 Comparativa
        campo = pedir_texto("  Campo para lineal (id/producto/vendedor): ", max_len=20)
        valor = pedir_texto("  Valor: ", max_len=50)
        monto = pedir_float("  Monto para binaria ($): ", minimo=0)

        _, c_li = busqueda_lineal_iter(campo, valor)
        _, c_lr = busqueda_lineal_rec(campo, valor)
        lista_ord, _, _ = quicksort(_copiar_lista(), "total")
        _, c_bi = busqueda_binaria_iter(lista_ord, monto)
        _, c_br = busqueda_binaria_rec(lista_ord, monto)

        _separador()
        print(f"  {'Algoritmo':<28} {'Comparaciones':>14}")
        _separador("-")
        print(f"  {'Lineal Iterativa':<28} {c_li:>14}")
        print(f"  {'Lineal Recursiva':<28} {c_lr:>14}")
        print(f"  {'Binaria Iterativa':<28} {c_bi:>14}")
        print(f"  {'Binaria Recursiva':<28} {c_br:>14}")
        _separador()
        _pausa()


# =============================================================================
# MÓDULO 6 – REPORTES
# =============================================================================

def exportar_txt():
    """RF-25: Exporta ventas a ventas_reporte.txt."""
    nombre = "ventas_reporte.txt"
    ahora  = datetime.now().strftime("%d/%m/%Y %H:%M:%S")
    total  = suma_recursiva(ventas)

    with open(nombre, "w", encoding="utf-8") as f:
        f.write("=" * 72 + "\n")
        f.write("           TECHSTORE – REPORTE DE VENTAS\n")
        f.write(f"           Generado: {ahora}\n")
        f.write("=" * 72 + "\n\n")
        f.write(f"{'ID':>4}  {'Producto':<22} {'Cant':>4} {'Precio':>12} "
                f"{'Total':>14}  {'Fecha':<12} {'Vendedor':<18} {'Sucursal'}\n")
        f.write("-" * 72 + "\n")
        for v in ventas:
            f.write(f"{v['id']:>4}  {v['producto']:<22} {v['cantidad']:>4}  "
                    f"${v['precio']:>11,.0f}  ${v['total']:>13,.0f}  "
                    f"{v['fecha']:<12} {v['vendedor']:<18} {v['sucursal']}\n")
        f.write("-" * 72 + "\n")
        f.write(f"\n  Total de registros : {len(ventas)}\n")
        f.write(f"  TOTAL ACUMULADO    : ${total:,.0f}\n")
        f.write("\n" + "=" * 72 + "\n")

    print(f"  ✔ Reporte exportado → {os.path.abspath(nombre)}")
    _pausa()

    from google.colab import files
    files.download("ventas_reporte.txt")
    files.download("ventas.csv")

def exportar_csv():
    """RF-26: Exporta ventas a ventas.csv compatible con Excel."""
    nombre = "ventas.csv"
    campos = ["id", "producto", "cantidad", "precio", "total",
              "fecha", "vendedor", "sucursal"]
    with open(nombre, "w", newline="", encoding="utf-8-sig") as f:
        writer = csv.DictWriter(f, fieldnames=campos)
        writer.writeheader()
        writer.writerows(ventas)
    print(f"  ✔ CSV exportado → {os.path.abspath(nombre)}")
    _pausa()

    from google.colab import files
    files.download("ventas_reporte.txt")
    files.download("ventas.csv")

def menu_reportes():
    _titulo("MÓDULO DE REPORTES")
    print("  1. Exportar TXT  (ventas_reporte.txt)")
    print("  2. Exportar CSV  (ventas.csv)")
    print("  3. Ambos formatos")
    opc = pedir_int("  Elige (1-3): ", minimo=1)
    if opc in (1, 3):
        exportar_txt()
    if opc in (2, 3):
        exportar_csv()


# =============================================================================
# MENÚ PRINCIPAL
# =============================================================================

def menu_gestion():
    while True:
        _titulo("MÓDULO 1 – GESTIÓN DE REGISTROS")
        print("  1. Ingresar nueva venta")
        print("  2. Mostrar todos los registros")
        print("  3. Modificar venta por ID")
        print("  4. Eliminar venta por ID")
        print("  0. Volver al menú principal")
        opc = input("  Opción: ").strip()
        if opc == "1":
            ingresar_venta()
        elif opc == "2":
            mostrar_ventas()
        elif opc == "3":
            modificar_venta()
        elif opc == "4":
            eliminar_venta()
        elif opc == "0":
            break
        else:
            print("  ✗ Opción inválida.")


def main():
    os.system("cls" if os.name == "nt" else "clear")
    _titulo("TECHSTORE – SISTEMA DE GESTIÓN DE VENTAS")
    print("  Cargando datos de ejemplo...")
    cargar_datos_ejemplo()
    _pausa()

    while True:
        os.system("cls" if os.name == "nt" else "clear")
        _titulo("MENÚ PRINCIPAL")
        print("  1. Gestión de Registros")
        print("  2. Análisis Estadístico")
        print("  3. Ordenamiento")
        print("  4. Búsqueda")
        print("  5. Recursividad (demostración)")
        print("  6. Reportes (TXT / CSV)")
        print("  0. Salir")
        _separador()
        opc = input("  Opción: ").strip()

        if opc == "1":
            menu_gestion()
        elif opc == "2":
            calcular_estadisticas()
        elif opc == "3":
            menu_ordenamiento()
        elif opc == "4":
            menu_busqueda()
        elif opc == "5":
            demo_recursividad()
        elif opc == "6":
            menu_reportes()
        elif opc == "0":
            print("\n  ¡Hasta luego! – TechStore 2024")
            break
        else:
            print("  ✗ Opción inválida.")
            _pausa()


# =============================================================================
if __name__ == "__main__":
    main()
```
